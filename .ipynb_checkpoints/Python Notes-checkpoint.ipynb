# K-means Clustering

```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
```

```python
myArr = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [4, 2],
    [4, 4],
    [4, 0]
])
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(myArr)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

# BeautifulSoup

```python
from bs4 import BeautifulSoup
import urllib
```

```python
r = urllib.request.Request('https://aflcio.org/what-unions-do/social-economic-justice/advocacy/legislative-alerts', headers={
    'User-Agent' : "Magic Browser"
})
```

```python
con = urllib.request.urlopen(r)
text = (con.read())
type(text)
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
logs = {}
```

```python
alerts = soup.find_all('a', class_='b-inner')
prefix = 'https://www.aflcio.org'
for a in alerts:
#     print(a.get_text())
#     print(prefix + a['href'])
    logs[a.get_text()] = {}
    logs[a.get_text()]['url'] = prefix + a['href']
    logs[a.get_text()]['text'] = get_link_text(a['href'])
```

```python
def get_link_text(link):
    new = urllib.request.urlopen(urllib.request.Request('https://aflcio.org'+link, headers={
        'User-Agent' : "Magic Browser"
    })).read()
    newsoup = BeautifulSoup(new, 'html.parser')
    finds = newsoup.find_all('div', id='block-afl-content')
    return finds[0].get_text()
```

```python
get_link_text(alerts[0]['href'])
```
